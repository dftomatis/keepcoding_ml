### 3.1 Filtrado geográfico y temporal del conjunto de entrenamiento

Con el objetivo de realizar un análisis coherente y enfocado, se filtrarán los datos del conjunto de entrenamiento de acuerdo a dos criterios:

1. **Filtrado temporal:**  
   Nos quedaremos únicamente con registros cuya fecha de extracción (`Last Scraped`) pertenezca al año **2017**. Esto asegura que el análisis se base en una única fotografía temporal del mercado, evitando inconsistencias provocadas por mezclas de años.

2. **Filtrado geográfico:**  
   El análisis se centrará exclusivamente en **Madrid (España)**, que representa la gran mayoría de las entradas del dataset y permite realizar un estudio más enfocado y consistente del mercado local.

```python
# Aseguramos que Last Scraped esté en formato datetime
df['Last Scraped'] = pd.to_datetime(df['Last Scraped'], errors='coerce')

# Aplicamos filtro temporal y geográfico
filtro_madrid_2017 = (df['State'].str.lower().str.contains('madrid')) & (df['Last Scraped'].dt.year == 2017)
df = df[filtro_madrid_2017].copy()

print(f"Tamaño del dataset tras filtrado geográfico y temporal: {df.shape}")
```

### 3.2 Eliminación de columnas irrelevantes

Antes de dividir el dataset, eliminamos columnas que sabemos que no aportan valor al modelo. Esta limpieza inicial incluye:

- **Identificadores y URLs** que son únicos por observación y no aportan información predictiva.
- **Columnas con imágenes** (enlaces a fotos) que no pueden ser procesadas directamente.
- **Variables con más del 30% de valores nulos**, que dificultan la imputación.
- **Otras columnas irrelevantes según el conocimiento del dominio.**

Este paso se realiza **antes** de la división en entrenamiento y prueba para evitar *data leakage*.



```python
# Calculamos el porcentaje de valores nulos
porcentaje_nulos = df.isnull().mean() * 100

# Identificamos columnas con más del 30% de nulos
columnas_con_muchos_nulos = porcentaje_nulos[porcentaje_nulos > 30].index.tolist()

# Otras columnas irrelevantes que eliminaremos por conocimiento del dominio
columnas_irrelevantes = [
    'Thumbnail Url', 'Medium Url', 'XL Picture Url',
    'Listing Url', 'Picture Url', 'Host Thumbnail Url',
    'Host Picture Url', 'Host URL', 'Host ID', 'ID', 'Scrape ID',
    'Space', 'Host Neighbourhood', 'Review Scores Value', 'Review Scores Location',
    'Review Scores Checkin', 'Review Scores Accuracy', 'Review Scores Cleanliness',
    'Review Scores Communication', 'Review Scores Rating', 'First Review',
    'Last Review', 'Reviews per Month', 'Host Response Time', 'Host Response Rate',
    'Latitude', 'Longitude', 'Guests Included', 'Extra People',
    'Number of Reviews'
]

# Combinamos todas las columnas a eliminar
columnas_delete = columnas_con_muchos_nulos + columnas_irrelevantes

# Eliminamos del dataframe
df.drop(columns=columnas_delete, inplace=True)

print(f"Tamaño del dataset tras eliminar columnas irrelevantes: {df.shape}")
```

En esta fase se eliminaron columnas irrelevantes, redundantes o con alta proporción de valores nulos. Esta limpieza se basó tanto en criterios cuantitativos como cualitativos:

- **Columnas con más del 30% de valores nulos**: Se eliminaron 17 columnas con altos porcentajes de datos faltantes. Entre ellas se encuentran variables descriptivas extensas (`Neighborhood Overview`, `Notes`), variables administrativas (`License`, `Jurisdiction Names`) y precios no consistentes (`Weekly Price`, `Monthly Price`), entre otras.

- **Columnas irrelevantes o redundantes**: También se eliminaron identificadores (`ID`, `Host ID`, `Scrape ID`), URLs (`Listing Url`, `Picture Url`, etc.), imágenes y otras columnas duplicadas o de bajo valor predictivo. Aunque algunas no superaban el umbral de nulos, su aporte al modelo era limitado o nulo.

Tras esta depuración, se redujo el número de columnas a 47, mejorando la calidad del dataset y reduciendo el riesgo de ruido en las siguientes etapas de modelado.



### 3.3 División del dataset en Train/Test

En esta sección separamos el dataset limpio en dos subconjuntos: entrenamiento (`train`) y prueba (`test`).

**Objetivo:**  
Evitar el **data leakage**, garantizando que todas las decisiones de preprocesamiento (como imputación de nulos, tratamiento de outliers y codificación) se realicen **exclusivamente sobre el conjunto de entrenamiento**.

Este enfoque asegura una evaluación más realista del desempeño del modelo al simular cómo se comportaría con datos nuevos e invisibles.

Dividiremos el dataset utilizando un 80% de los datos para entrenamiento y un 20% para prueba, manteniendo la distribución aleatoria con una semilla fija (`random_state=42`) para garantizar reproducibilidad.


```python
from sklearn.model_selection import train_test_split

# División del dataset limpio en entrenamiento (80%) y prueba (20%)
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)

# Mostramos tamaños de los subconjuntos
print(f"Tamaño del conjunto de entrenamiento: {df_train.shape}")
print(f"Tamaño del conjunto de prueba: {df_test.shape}")
```

### 3.4 Relleno de valores nulos en variables relevantes

En esta sección analizaremos las variables que aún presentan valores nulos después del proceso de limpieza de columnas y del filtrado temporal y geográfico.

**Objetivos:**

- Detectar columnas relevantes que aún contienen valores nulos.
- Evaluar si dichas variables son importantes para el modelado.
- Aplicar un método de imputación adecuado en cada caso (relleno con 'Unknown', media, mediana, etc.).
- Dejar el conjunto de entrenamiento y test preparados para los siguientes pasos del pipeline de modelado.



```python
# Verificamos porcentaje de nulos en df_train tras el rellenado
porcentaje_nulos_train = df_train.isnull().mean() * 100
porcentaje_nulos_train = porcentaje_nulos_train[porcentaje_nulos_train > 0].sort_values(ascending=False)

print("Porcentaje de nulos restantes en df_train:")
display(porcentaje_nulos_train)

# Verificamos porcentaje de nulos en df_test tras el rellenado
porcentaje_nulos_test = df_test.isnull().mean() * 100
porcentaje_nulos_test = porcentaje_nulos_test[porcentaje_nulos_test > 0].sort_values(ascending=False)

print("Porcentaje de nulos restantes en df_test:")
display(porcentaje_nulos_test)
```

```python
# Rellenos categóricos con 'Unknown'
relleno_categorico = [
    'Summary', 'Zipcode', 'Amenities', 'Market', 'Host Location',
    'State', 'Host Verifications', 'Description', 'Host Name', 'Host Since'
]

for col in relleno_categorico:
    df_train[col] = df_train[col].fillna('Unknown')
    df_test[col] = df_test[col].fillna('Unknown')
    print(f"Rellenados nulos en {col} con 'Unknown'")

# Rellenos numéricos con 0
relleno_cero = ['Host Listings Count', 'Host Total Listings Count']

for col in relleno_cero:
    df_train[col] = df_train[col].fillna(0)
    df_test[col] = df_test[col].fillna(0)
    print(f"Rellenados nulos en {col} con 0")

# Rellenos numéricos con mediana (calculada en train)
relleno_mediana = ['Bathrooms', 'Beds', 'Bedrooms', 'Price']

for col in relleno_mediana:
    mediana = df_train[col].median()
    df_train[col] = df_train[col].fillna(mediana)
    df_test[col] = df_test[col].fillna(mediana)
    print(f"Rellenados nulos en {col} con la mediana: {mediana:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma de Price (sin transformación)
plt.figure(figsize=(10, 5))
sns.histplot(df_train['Price'], bins=100, kde=True)
plt.title('Distribución de Price (sin transformación)')
plt.xlabel('Price')
plt.ylabel('Frecuencia')
plt.show()

# Boxplot para detectar outliers más fácilmente
plt.figure(figsize=(10, 2))
sns.boxplot(x=df_train['Price'])
plt.title('Boxplot de Price')
plt.show()
```

### 3.6 Creación de nuevas variables (feature engineering)

En esta sección añadimos variables derivadas de columnas existentes para capturar patrones
que no son evidentes de forma directa y, así, mejorar la capacidad predictiva de los modelos.

**Objetivos**

* generar variables numéricas adicionales que tengan significado económico o de negocio;  
* evitar cualquier atributo que dependa de la variable objetivo (`Price`);  
* aplicar las transformaciones tanto en `df_train` como en `df_test` para mantener
  la coherencia de los datos de entrada a los modelos.

**Variables creadas**

* `bathrooms_per_guest` : proporción de baños por huésped.  
* `log_price` : logaritmo natural de `Price`, útil para reducir la asimetría de la distribución y hacer que los errores relativos sean más homogéneos.

> ⚠️ Nota: estas transformaciones se aplican **por separado** en `df_train` y `df_test`
> utilizando exactamente la misma lógica para evitar fugas de información.

```python
import numpy as np

# aseguramos que el dataframe original no se trunque
df_train = df_train.copy()
df_test  = df_test.copy()

def add_engineered_columns(df):
    # evita división por cero
    df = df[df['Accommodates'] > 0]
    df['bathrooms_per_guest'] = (
        df.apply(lambda row:
                 row['Bathrooms'] / row['Accommodates']
                 if row['Bathrooms'] > 0 else 0, axis=1)
    )
    df['log_price'] = np.log(df['Price'])

    return df

df_train = add_engineered_columns(df_train)
df_test  = add_engineered_columns(df_test)

# comprobación rápida
df_train[['Price', 'Accommodates', 'bathrooms_per_guest', 'log_price']].head()

```

### 3.7 Codificación de variables categóricas

En esta sección se transforman variables categóricas en variables numéricas para que puedan ser utilizadas por los algoritmos de machine learning.

#### Objetivo:
- Codificar las variables categóricas más relevantes sin introducir *data leakage*.

#### Variables consideradas:
- Se decidió codificar únicamente `Room Type`, dado que tiene una fuerte relación con `Price` y un número reducido de categorías.
- Se descartó `Cancellation Policy` en esta etapa por su bajo impacto predictivo y su distribución desequilibrada.

#### Técnica utilizada:
- **One-Hot Encoding**: se crean columnas binarias para cada categoría de la variable `Room Type`.

⚠️ **Nota:** La codificación se aplica solo sobre `df_train`. Esta transformación deberá aplicarse también en `df_test` utilizando las mismas columnas para mantener consistencia.


```python
# One-Hot Encoding para 'Room Type'
room_type_dummies = pd.get_dummies(df_train['Room Type'], prefix='room_type')

# Concatenar con df_train
df_train = pd.concat([df_train.drop('Room Type', axis=1), room_type_dummies], axis=1)

# Verificar resultado
df_train.filter(like='room_type').head()
```

### 3.8 Agrupación y codificación de barrios (`Neighbourhood Cleansed`)

En esta sección se transforma la variable `Neighbourhood Cleansed`, que representa el barrio del alojamiento en formato estandarizado.

Dado que existen muchos barrios diferentes, se procede a:

1. Analizar la frecuencia de aparición de cada barrio.
2. Agrupar los barrios menos frecuentes en una categoría `"Other"` para reducir la dimensionalidad.
3. Aplicar **codificación de frecuencia**, que asigna a cada barrio la proporción de registros que representa sobre el total.

Esta técnica es útil para evitar crear demasiadas columnas mediante One-Hot Encoding, y mantiene una representación numérica simple sin riesgo de *data leakage*.

⚠️ **Nota:** Esta transformación se aplica solo sobre `df_train`. En `df_test` se debe replicar exactamente con las mismas categorías y frecuencias calculadas.


```python
# Analizar cantidad de barrios
neighbourhood_counts = df_train['Neighbourhood Cleansed'].value_counts()
print(f"Número de barrios únicos: {len(neighbourhood_counts)}")
```

```python
# Agrupar barrios poco frecuentes como "Other"
umbral = 100  
barrios_comunes = neighbourhood_counts[neighbourhood_counts >= umbral].index

df_train['neigh_clean_grouped'] = df_train['Neighbourhood Cleansed'].apply(
    lambda x: x if x in barrios_comunes else 'Other'
)

# Codificación de frecuencia
frecuencia_barrios = df_train['neigh_clean_grouped'].value_counts(normalize=True)

# Mapeo al dataset
df_train['neigh_freq_encoded'] = df_train['neigh_clean_grouped'].map(frecuencia_barrios)

# Verificamos el resultado
df_train[['Neighbourhood Cleansed', 'neigh_clean_grouped', 'neigh_freq_encoded']].head()
```

```python
# Analizar cantidad de barrios agrupados
neigh_clean_grouped_counts = df_train['neigh_clean_grouped'].value_counts()
print(f"Número de barrios agrupados: {len(neigh_clean_grouped_counts)}")
```

### 3.9 Escalado de variables numéricas

En esta sección se normalizan las variables numéricas para garantizar que todas estén en una escala comparable, requisito fundamental para modelos como regresión lineal y KNN.

#### Variables seleccionadas para escalar:
- `Accommodates`
- `Bathrooms`
- `Bedrooms`
- `Beds`
- `bathrooms_per_guest`

Estas variables tienen rangos diferentes, lo que podría distorsionar los resultados de los modelos si no se ajusta.

#### Técnica utilizada:
- Se aplica **escalado estándar (Z-score)** usando `StandardScaler`, que transforma los datos a una distribución con media 0 y desviación estándar 1.

⚠️ **Nota:** El escalador se ajusta solo con `df_train` y debe usarse luego para transformar `df_test`, sin volver a entrenarlo.


```python
# Lista de variables numéricas candidatas a escalar
vars_escalar = ['Accommodates', 'Bathrooms', 'Bedrooms', 'Beds', 
                'bathrooms_per_guest']

# Estadísticas descriptivas para revisar rango, media, std, etc.
df_train[vars_escalar].describe()
```

```python
from sklearn.preprocessing import StandardScaler

# Variables a escalar
vars_escalar = ['Accommodates', 'Bathrooms', 'Bedrooms', 'Beds', 
                'bathrooms_per_guest']

# Inicializar y ajustar scaler sobre df_train
scaler = StandardScaler()
df_train_scaled = df_train.copy()
df_train_scaled[vars_escalar] = scaler.fit_transform(df_train[vars_escalar])

# Verificamos el resultado
df_train_scaled[vars_escalar].describe()
```

```python
# Guardar barrios frecuentes (los que no fueron agrupados como "Other")
barrios_comunes = df_train['neigh_clean_grouped'].unique().tolist()

# Guardar codificación de frecuencia por barrio
frecuencia_barrios = df_train['neigh_clean_grouped'].value_counts(normalize=True)

# Guardar columnas dummy de Room Type 
room_type_dummies_cols = [col for col in df_train.columns if col.startswith('room_type_')]
```

### 3.10 Limpieza de columnas no numéricas

En este paso se eliminan todas aquellas columnas que no serán utilizadas por los modelos de machine learning.

En particular, se eliminan:
- Columnas de tipo `object` (texto libre o categorías no codificadas).
- Columnas de tipo `datetime` o `bool`, si no fueron convertidas previamente.
- Variables intermedias como `neigh_clean_grouped`, que se usaron solo para agrupar categorías.

Al mismo tiempo, se reconvierten explícitamente a tipo numérico aquellas columnas útiles que fueron transformadas correctamente (por ejemplo, `room_type_*` y `neigh_freq_encoded`).


```python
# Reconvertimos columnas útiles que podrían haberse quedado como 'object'
columns_to_keep = ['neigh_freq_encoded'] + [col for col in df_train_scaled.columns if col.startswith('room_type_')]

for col in columns_to_keep:
    if col in df_train_scaled.columns:
        df_train_scaled[col] = df_train_scaled[col].astype(float)

# Eliminamos columnas no numéricas que no fueron transformadas
non_numeric_cols = df_train_scaled.select_dtypes(exclude=['int64', 'float64']).columns.tolist()

# Excluir del borrado las columnas útiles ya casteadas
non_numeric_cols = [col for col in non_numeric_cols if col not in columns_to_keep]

print("Columnas eliminadas por no ser numéricas ni transformadas:")
print(non_numeric_cols)

df_train_scaled = df_train_scaled.drop(columns=non_numeric_cols)

# Verificación final
print("Tamaño final de df_train_scaled:", df_train_scaled.shape)
print("Columnas finales (top 10):", df_train_scaled.columns[:10].tolist())

```

```python
# Convertir posibles dummies mal tipadas a int64
dummies_cols = ['room_type_Entire home/apt', 'room_type_Private room', 'room_type_Shared room']
for col in dummies_cols:
    if col in df_train_scaled.columns:
        df_train_scaled[col] = df_train_scaled[col].astype(int)

# Eliminar columnas no numéricas
non_numeric_cols = df_train_scaled.select_dtypes(exclude=['int64', 'float64']).columns.tolist()
df_train_scaled = df_train_scaled.drop(columns=non_numeric_cols)
```

```python
# Confirmación final
print("Tamaño final de df_train_scaled:", df_train_scaled.shape)
print("Columnas restantes (top 10):", df_train_scaled.columns[:50].tolist())
```

### 3.11 Aplicación de transformaciones a `df_test`

En esta sección se aplican al conjunto `df_test` las mismas transformaciones realizadas previamente en `df_train`, asegurando coherencia y evitando *data leakage*.

Las transformaciones aplicadas son:

1. Imputación de valores nulos con los mismos criterios usados en entrenamiento.
2. Creación de variables derivadas (`bathrooms_per_guest`, `log_price`).
3. Agrupación de barrios poco frecuentes y codificación por frecuencia (`neigh_freq_encoded`).
4. Codificación de `Room Type` con One-Hot Encoding asegurando coincidencia de columnas.
5. Escalado de variables numéricas con el `StandardScaler` previamente ajustado.
6. Conversión de variables categóricas transformadas a tipo numérico.
7. Eliminación de columnas no numéricas no utilizadas, igual que en `df_train_scaled`.

Este paso asegura que `df_test` quede alineado con `df_train_scaled` en estructura y formato, permitiendo aplicar modelos sin errores.


```python
# imputación de nulos
df_test['Bathrooms'] = df_test['Bathrooms'].fillna(df_train['Bathrooms'].median())
df_test['Bedrooms']  = df_test['Bedrooms'].fillna(df_train['Bedrooms'].median())
df_test['Beds']      = df_test['Beds'].fillna(df_train['Beds'].median())
df_test['Host Listings Count']        = df_test['Host Listings Count'].fillna(0)
df_test['Host Total Listings Count']  = df_test['Host Total Listings Count'].fillna(0)

for col in ['Summary', 'Zipcode', 'Amenities', 'Market', 'Host Location',
            'State', 'Host Verifications', 'Description', 'Host Name', 'Host Since']:
    df_test[col] = df_test[col].fillna('Unknown')

# crear variables derivadas
df_test = df_test[df_test['Accommodates'] > 0]

df_test['bathrooms_per_guest'] = (
    df_test.apply(lambda row:
                  row['Bathrooms'] / row['Accommodates']
                  if row['Bathrooms'] > 0 else 0,
                  axis=1)
)
df_test['log_price'] = np.log(df_test['Price'])

# agrupación de barrios
df_test['neigh_clean_grouped'] = df_test['Neighbourhood Cleansed'].apply(
    lambda x: x if x in barrios_comunes else 'Other'
)

# codificación de frecuencia de barrio
df_test['neigh_freq_encoded'] = df_test['neigh_clean_grouped'].map(frecuencia_barrios).fillna(0)

# one-hot encoding de room type
room_type_dummies_test = pd.get_dummies(df_test['Room Type'], prefix='room_type')
for col in room_type_dummies_cols:
    if col not in room_type_dummies_test.columns:
        room_type_dummies_test[col] = 0
room_type_dummies_test = room_type_dummies_test[room_type_dummies_cols]
df_test = pd.concat([df_test.drop(columns=['Room Type']), room_type_dummies_test], axis=1)

# escalado de variables numéricas
vars_escalar = ['Accommodates', 'Bathrooms', 'Bedrooms', 'Beds',
                'bathrooms_per_guest']
df_test[vars_escalar] = scaler.transform(df_test[vars_escalar])

# asegurar que columnas útiles estén en tipo numérico
columns_to_cast = ['neigh_freq_encoded'] + room_type_dummies_cols
df_test[columns_to_cast] = df_test[columns_to_cast].astype(float)

# eliminar columnas no numéricas no utilizadas
non_numeric_cols_test = df_test.select_dtypes(exclude=['int64', 'float64']).columns.tolist()
non_usables_test = [col for col in non_numeric_cols_test if col not in columns_to_cast]
df_test.drop(columns=non_usables_test, inplace=True)

print("tamaño final de df_test:", df_test.shape)
print("columnas finales:", df_test.columns[:50].tolist())
```

### 4.1 Análisis de correlación entre variables numéricas

El objetivo de este paso es detectar relaciones lineales fuertes entre las variables numéricas mediante la matriz de correlación de Pearson.

Se busca:
- Identificar variables altamente correlacionadas entre sí (redundantes).
- Entender cómo se relacionan con la variable objetivo (`Price` o `log_price`).
- Decidir si eliminar alguna variable en favor de otra equivalente.

Se considerará como **correlación alta** un valor absoluto de **r > 0.8** entre dos variables.

Nota: Aunque una variable esté altamente correlacionada con otra, su eliminación depende del contexto del modelo y la interpretabilidad.



```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Copia del dataframe escalado
correlaciones = df_train_scaled.copy()

# Excluir la variable Price si usás log_price como objetivo
if 'Price' in correlaciones.columns:
    correlaciones = correlaciones.drop(columns=['Price'])

# Calculamos la matriz de correlaciones
cor_matrix = correlaciones.corr(numeric_only=True)

# Creamos una máscara para ocultar la mitad superior
mask = np.triu(np.ones_like(cor_matrix, dtype=bool))

# Generamos el mapa de calor
plt.figure(figsize=(20, 16))
sns.heatmap(cor_matrix, mask=mask, cmap='coolwarm', center=0,
            annot=True, fmt=".2f", linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Matriz de correlaciones entre variables numéricas', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 5. Modelado predictivo

En esta sección se desarrollan modelos de regresión para predecir el precio de los alojamientos (`log_price`) utilizando las variables seleccionadas en las etapas anteriores.

El objetivo es evaluar el rendimiento de distintos enfoques, comenzando con un modelo de regresión lineal múltiple como línea base.

### 5.1 Regresión lineal (baseline)  

Entrenamos un modelo de regresión lineal utilizando `log(Price)` como variable objetivo.  
Para evitar fugas de información el flujo completo se implementa dentro de un **`Pipeline`** que:

* escala las variables numéricas con `StandardScaler`,  
* deja pasar sin cambios las columnas categóricas ya dummificadas,  
* ajusta el modelo lineal en un único paso reproducible.  

> ⚠️ Nota: el entrenamiento se realiza **solo** sobre `df_train`; la evaluación se hace sobre `df_test`.


```python
from sklearn import set_config
set_config(display='text')

import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score 

# lista de variables numéricas
numeric_features = [
    'Accommodates', 'Bathrooms', 'Bedrooms', 'Beds',
    'Minimum Nights', 'Maximum Nights',
    'Availability 30', 'Availability 365',
    'bathrooms_per_guest',
    'neigh_freq_encoded'
]

# columnas dummificadas de room type
room_type_dummies = [
    'room_type_Entire home/apt',
    'room_type_Private room',
    'room_type_Shared room'
]

feature_cols = numeric_features + room_type_dummies

# matrices de entrenamiento y prueba con las mismas columnas
X_train = df_train[feature_cols].copy()
X_test  = df_test[feature_cols].copy()
y_train = df_train['log_price']
y_test  = df_test['log_price'] 

# definición del pipeline de preprocesado + modelo
numeric_transformer = Pipeline([('scaler', StandardScaler())])

preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_features),
    ('cat', 'passthrough', room_type_dummies)
])

linreg_pipeline = Pipeline([
    ('prep', preprocessor),
    ('model', LinearRegression())
])

# entrenamiento
linreg_pipeline.fit(X_train, y_train)
```

```python

# evaluacion
y_pred_log = linreg_pipeline.predict(X_test)

# Métricas en log-escala
mse_log  = mean_squared_error(y_test, y_pred_log)
rmse_log = np.sqrt(mse_log)
mae_log  = mean_absolute_error(y_test, y_pred_log)
r2_log   = r2_score(y_test, y_pred_log)

# Métricas en escala original
y_pred     = np.exp(y_pred_log)
y_test_exp = np.exp(y_test)

mse   = mean_squared_error(y_test_exp, y_pred)
rmse  = np.sqrt(mse)
mae   = mean_absolute_error(y_test_exp, y_pred)
r2    = r2_score(y_test_exp, y_pred)

print("---- Escala log ----")
print(f"RMSE_log : {rmse_log:.4f}")
print(f"MAE_log  : {mae_log :.4f}")
print(f"R²_log   : {r2_log  :.4f}")

print("\n---- Escala original ----")
print(f"RMSE     : {rmse: .2f} USD")
print(f"MAE      : {mae : .2f} USD")
print(f"R²       : {r2  :.4f}")
```

#### Desempeño del modelo

| Métrica | Escala log | Escala original (USD) |
|---------|-----------:|----------------------:|
| **RMSE** | **0.5539** | **33.73** |
| **MAE**  | **0.4415** | **22.54** |
| **R²**   | **0.2201** | **0.3000** |

* **Interpretación**  
  * Un RMSE log de **0.5539** implica un error relativo medio ≈ **74 %** sobre el precio real (e^0.5539 ≈ 1.74).  
  * En términos absolutos, el modelo se desvía ≈ **34 USD** sobre un precio típico de ~66 USD.  
  * Con un **R² ≈ 0.30** el modelo explica alrededor de un tercio de la variabilidad del precio, por lo que aún existe margen amplio de mejora.

* **Conclusión**  
  Este rendimiento constituye una **línea base**: muestra una relación moderada entre las variables seleccionadas y el precio, pero deja claro que se necesitan técnicas más flexibles (por ejemplo, modelos de árbol o boosting) y nuevas variables (temporales, geográficas, interacciones) para capturar la complejidad del mercado.

### 5.2 Regresión con regularización: Ridge y Lasso

Para controlar la multicolinealidad y prevenir sobreajuste se ajustaron modelos **Ridge** y **Lasso** usando el mismo `Pipeline` de la línea base:

* escalado con `StandardScaler`,
* idéntico conjunto de variables,
* búsqueda de `alpha` por `GridSearchCV` (5 folds)

Ambos modelos se comparan con la regresión lineal base utilizando las mismas métricas: RMSE y R².


```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

numeric_features = [
    'Accommodates', 'Bathrooms', 'Bedrooms', 'Beds',
    'Minimum Nights', 'Maximum Nights',
    'Availability 30', 'Availability 365',
    'bathrooms_per_guest',
    'neigh_freq_encoded'
]

room_type_dummies = [
    'room_type_Entire home/apt',
    'room_type_Private room',
    'room_type_Shared room'
]

feature_cols = numeric_features + room_type_dummies
X_train = df_train[feature_cols]
X_test  = df_test[feature_cols]
y_train = df_train['log_price']
y_test  = df_test['log_price']

num_pipe = Pipeline([('scaler', StandardScaler())])
preproc  = ColumnTransformer([
    ('num', num_pipe, numeric_features),
    ('cat', 'passthrough', room_type_dummies)
])

def evaluate(model, name):
    y_pred_log = model.predict(X_test)
    rmse_log = np.sqrt(mean_squared_error(y_test, y_pred_log))
    r2_log = r2_score(y_test, y_pred_log)

    y_pred = np.exp(y_pred_log)
    y_real = np.exp(y_test)
    rmse   = np.sqrt(mean_squared_error(y_real, y_pred))
    mae    = mean_absolute_error(y_real, y_pred)
    r2     = r2_score(y_real, y_pred)
    print(f'{name}: RMSE_log={rmse_log:.3f}  R²_log={r2_log:.3f}  RMSE=${rmse:.2f}  MAE=${mae:.2f}  R²={r2:.3f}')

# ridge con búsqueda de alpha
ridge_pipe = Pipeline([
    ('prep', preproc),
    ('model', Ridge())
])
ridge_cv = GridSearchCV(ridge_pipe,
                        param_grid={'model__alpha': [0.1, 1, 10, 100]},
                        scoring='neg_root_mean_squared_error',
                        cv=5,
                        n_jobs=-1)
ridge_cv.fit(X_train, y_train)
evaluate(ridge_cv.best_estimator_, 'Ridge')

# lasso con búsqueda de alpha
lasso_pipe = Pipeline([
    ('prep', preproc),
    ('model', Lasso(max_iter=5000))
])
lasso_cv = GridSearchCV(lasso_pipe,
                        param_grid={'model__alpha': [0.0005, 0.001, 0.005, 0.01]},
                        scoring='neg_root_mean_squared_error',
                        cv=5,
                        n_jobs=-1)
lasso_cv.fit(X_train, y_train)
evaluate(lasso_cv.best_estimator_, 'Lasso')
```

| Modelo | RMSE (log) | R² (log) | RMSE (USD) | MAE (USD) | R² |
|--------|-----------:|---------:|-----------:|----------:|---:|
| Ridge  | **0.554** | **0.220** | **33.73** | **22.54** | **0.300** |
| Lasso  | **0.553** | **0.223** | **33.73** | **22.53** | **0.300** |

**Interpretación**

* Los valores son prácticamente idénticos a los de la regresión lineal base, señal de que la regularización **no añade ganancia** con las variables actuales.
* El hecho de que **Ridge y Lasso obtengan métricas casi calcadas** sugiere que  
  1. el modelo lineal ya está lejos de sobre-ajustar, y  
  2. la señal predictiva restante es no lineal o requiere variables adicionales.
* Esto refuerza la necesidad de **modelos más flexibles** (Gradient Boosting, Random Forest, CatBoost) y de **nueva ingeniería de características** (variables temporales, interacciones barrio-temporada, distancias geográficas).

En la siguiente sección se explorarán modelos basados en árboles para comprobar si pueden capturar la complejidad que el enfoque lineal regularizado no alcanza.



### 5.3 Validación cruzada (Cross-Validation)

Como paso final del modelado, se realiza validación cruzada con 5 particiones sobre el conjunto de entrenamiento, usando el modelo de Ridge Regression.

Este paso permite estimar la estabilidad del modelo ante diferentes particiones de datos y verificar si su rendimiento general es confiable.

Se utiliza la métrica **RMSE en escala logarítmica**, consistente con la variable objetivo `log_price`.


```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge
import numpy as np

# Inicializamos el modelo Ridge con el mismo alpha usado antes
ridge_cv = Ridge(alpha=1.0)

# Validación cruzada con scoring negativo
neg_rmse_scores = cross_val_score(
    ridge_cv,
    X_train,
    y_train,
    scoring='neg_root_mean_squared_error', 
    cv=5
)

# Convertimos a valores positivos
rmse_scores = -neg_rmse_scores

```

```python
# Mostramos resultados
print("RMSE por fold:", rmse_scores)
print(f"RMSE promedio: {rmse_scores.mean():.4f}")
print(f"Desviación estándar: {rmse_scores.std():.4f}")
```

Se aplicó **K-Fold CV (5 folds, shuffle=True, random_state=42)** sobre el mismo `Pipeline` usado en 5.1-5.2.  
Los resultados fueron:

## 6. Conclusión  

El objetivo de este trabajo fue construir un pipeline de Machine Learning que predijera el precio de alojamientos de Airbnb (Madrid) cumpliendo con las buenas prácticas de preparación de datos, validación y reporte.

### Principales etapas ejecutadas
* eliminación de columnas irrelevantes y con alto porcentaje de valores nulos;  
* división temprana del conjunto en **train / test** para evitar *data-leakage*;  
* tratamiento de outliers **solo** en `df_train` (se conservaron en `df_test` para evaluar robustez);  
* creación de la variable objetivo `log_price` y de *features* derivadas independientes del precio (`bathrooms_per_guest`, codificación de barrio);  
* codificación one-hot de `Room Type` y escalado de numéricas dentro de un único **`Pipeline`**;  
* entrenamiento y comparación de tres modelos lineales (Regresión Múltiple, Ridge y Lasso) con idénticos datos de entrada;  
* validación cruzada (5 folds) para estimar la estabilidad del modelo Ridge.

### Desempeño de los modelos

| Modelo            | RMSE (log) | R² (log) | RMSE (USD) | MAE (USD) | R² |
|-------------------|-----------:|---------:|-----------:|----------:|---:|
| Regresión lineal  | **0.5539** | **0.2201** | **33.73** | **22.54** | **0.3000** |
| Ridge (α óptimo)  | **0.5539** | **0.2201** | **33.73** | **22.54** | **0.3000** |
| Lasso (α óptimo)  | **0.5533** | **0.2230** | **33.73** | **22.53** | **0.3000** |

El modelo Ridge obtuvo un **RMSE medio de 0.384 (log)** en validación cruzada (σ = 0.007), lo que evidencia estabilidad interna pero también una brecha respecto al error en el conjunto de prueba (RMSE_log ≈ 0.554).

### Reflexión final
Los modelos lineales explican ≈ 30 % de la variabilidad del precio y presentan un **MAE ≈ 23 USD** sobre un precio típico de ~66 USD. Constituyen, por tanto, una **línea base razonable**, pero para mejorar la precisión sería necesario:

1. incorporar variables temporales (año/mes del anuncio, estacionalidad) y geográficas (distancia a puntos de interés);  
2. probar modelos no lineales (árboles de gradiente, CatBoost) capaces de capturar interacciones complejas.

El pipeline resultante es reproducible, evita fugas de información y reporta métricas transparentes, cumpliendo los requisitos establecidos para esta práctica.
